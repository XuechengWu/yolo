````
import os
import shutil

# srcfile éœ€è¦å¤åˆ¶ã€ç§»åŠ¨çš„æ–‡ä»¶   
# dstpath ç›®çš„åœ°å€
 
def mycopyfile(srcfile,dstpath):                       # å¤åˆ¶å‡½æ•°
    if not os.path.isfile(srcfile):
        print ("%s not exist!"%(srcfile))
    else:
        fpath,fname=os.path.split(srcfile)             # åˆ†ç¦»æ–‡ä»¶åå’Œè·¯å¾„
        if not os.path.exists(dstpath):
            os.makedirs(dstpath)                       # åˆ›å»ºè·¯å¾„
        shutil.copy(srcfile, dstpath + fname)          # å¤åˆ¶æ–‡ä»¶
        print ("copy %s -> %s"%(srcfile, dstpath + fname))
 
 
# src_dir = './'
````

````
import os
import zipfile
 
 
def zipDir(dirpath, outFullName):
    """
    åŽ‹ç¼©æŒ‡å®šæ–‡ä»¶å¤¹
    :param dirpath: ç›®æ ‡æ–‡ä»¶å¤¹è·¯å¾„
    :param outFullName: åŽ‹ç¼©æ–‡ä»¶ä¿å­˜è·¯å¾„+xxxx.zip
    :return: æ— 
    """
    zip = zipfile.ZipFile(outFullName, "w", zipfile.ZIP_DEFLATED)
    for path, dirnames, filenames in os.walk(dirpath):
        # åŽ»æŽ‰ç›®æ ‡è·Ÿè·¯å¾„ï¼Œåªå¯¹ç›®æ ‡æ–‡ä»¶å¤¹ä¸‹è¾¹çš„æ–‡ä»¶åŠæ–‡ä»¶å¤¹è¿›è¡ŒåŽ‹ç¼©
        fpath = path.replace(dirpath, '')
 
        for filename in filenames:
            zip.write(os.path.join(path, filename), os.path.join(fpath, filename))
    zip.close()
````
